```python
img1=cv2.imread("../../python for computer vision/data/dog_backpack.png",cv2.COLOR_BGR2RGB)

img2=cv2.imread("../../python for computer vision/data/watermark_no_copy.png",cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
plt.imshow(img2)
```

```python
img1=cv2.resize(img1,(1200,1200))
img2=cv2.resize(img2,(1200,1200))


```

```python
plt.imshow(img1)
```

```python
plt.imshow(img1)
```

```python
blanded=cv2.addWeighted(img1,0.7,img2,.7,0)
```

```python
plt.imshow(blanded)
```
